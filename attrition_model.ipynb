```python
df['Attrition']=df['Attrition'].map({'Yes':1,'No':0}).astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in df.select_dtypes(include='object').columns:
    df[i]=le.fit_transform(df[i])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
x=df.drop('Attrition',axis=1)
y=df['Attrition']
x_train, x_test , y_train , y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train= scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
input_data = np.array([[35, 2, 1100, 1, 10, 3, 4, 1, 1023, 3, 1, 70, 3, 2, 5, 3, 1, 50000, 15000, 2, 1, 1, 15, 3, 3, 80, 1, 10, 3, 3, 7, 4, 1, 5]])
input_scaled = scaler.transform(input_data)
prediction = model.predict(input_scaled)

input_scaled = scaler.transform(input_data)
prediction = model.predict(input_scaled)

if prediction[0] == 1:
    print("Employee is likely to LEAVE the company")
else:
    print("Employee is likely to STAY in the company")
```
